# Bollinger Bands - BB

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import pandas_ta as ta
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df = yf.download(tickers='BTC-USD', period='max', interval='1d', auto_adjust=False)
df.columns = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
df
```

```python
bbands = ta.bbands(df.Close, length=14, std=2)
df = df.join(bbands)
```

```python
df
```

```python
df = df[:500]

fig = go.Figure()

fig.add_trace(
    go.Candlestick(
        x = df.index,
        open = df['Open'],
        high = df['High'],
        low = df['Low'],
        close = df['Close'],
        name = 'OHLC'
    )
)

fig.add_trace(
    go.Scatter(
        x = df.index,
        y = df['BBU_14_2.0_2.0'],
        name = 'BB_UPPER'
    )
)

fig.add_trace(
    go.Scatter(
        x = df.index,
        y = df['BBM_14_2.0_2.0'],
        name = 'BB_MIDDLE'
    )
)

fig.add_trace(
    go.Scatter(
        x = df.index,
        y = df['BBL_14_2.0_2.0'],
        name = 'BB_LOWER'
    )
)

fig.update_layout(
    xaxis = dict(rangeslider = dict(visible=False))
)

fig.show()
```

# Average True Range - ATR

```python
df['ATR'] = ta.atr(df.High, df.Low, df.Close, length=14)
df
```

```python
# Create a subplot and position is 1 row x 2 columns
fig = make_subplots(rows=2, cols=1, subplot_titles=['Price', 'ATR'], shared_xaxes=True)

# In the first row, plot the candlestick chart
fig.add_trace(
    go.Candlestick(
        x = df.index,
        open = df['Open'],
        high = df['High'],
        low = df['Low'],
        close = df['Close'],
        name='OHLC'
    ),
    row = 1,
    col = 1

)

# In the second row, plot the ATR
fig.add_trace(
    go.Scatter(
        x = df.index,
        y = df['ATR'],
        name = 'ATR'   
    ),
    row = 2,
    col = 1
)

fig.update_layout(
    xaxis = dict(rangeslider = dict(visible=False))
)

fig.show()
```
